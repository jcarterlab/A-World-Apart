## 2) Get edges

```python
# gets a set of ranges in thousands for the target account's followers. 
def get_ranges(target_acc):
    total = target_acc.value_counts().sum()
    thousands = total / 1000

    ranges = []
    for i in range(int(thousands)):
        start = (i*10**3)
        end = (i*10**3) + (10**3)
        ranges.append(range(int(start), int(end)))
        
    return(ranges)

# gets the edges (connections) for a single range of the target account's followers. 
def get_range_edges(target_cc, test_acc, range_no):
    node = []
    edge = []
    for i in range(len(target_cc)):
        if np.isin(target_cc.user_id.iloc[i], test_acc.user_id):
            node.append(target_cc.party.iloc[i])
            edge.append(test_acc.user_id.loc[test_acc.user_id == target_cc.user_id.iloc[i]].values[0])
    
    d = {'node': node,'edge': edge, 'thousand': range_no}
    df = pd.DataFrame(d)
    
    return df

# gets the edges for all the ranges of the target account's follwers. 
def get_all_range_edges(target_acc, test_acc):
    ranges = get_ranges(target_acc)
    
    df = []
    for i in range(len(ranges)):
        df.append(get_range_edges(target_acc.iloc[ranges[i],], test_acc, range_no=i+1))

    return pd.concat(df)
```

## 3) Visualize ratios

```python
# imports packages. 
import networkx as nx
import pandas as pd
import numpy as np

# reads in the us edges data.
dems_edges = pd.read_excel('C:/Users/HUAWEI/Desktop/Projects/A-World-Apart/Data/dems.xlsx')
reps_edges = pd.read_excel('C:/Users/HUAWEI/Desktop/Projects/A-World-Apart/Data/reps.xlsx')

# reads in the uK edges data.
tories_edges = pd.read_excel('C:/Users/HUAWEI/Desktop/Projects/A-World-Apart/Data/tories.xlsx')
labour_edges = pd.read_excel('C:/Users/HUAWEI/Desktop/Projects/A-World-Apart/Data/labour.xlsx')

# views the first 4 rows of the dems edges data. 
dems_edges.head(4)
```

```python
# imports matplotlib. 
import matplotlib.pyplot as plt

# plots the raw data. 
def get_plot(names, colors = ['#0096FF', '#E35335'], label = ''):
    
    # creates variables for party no. and the no. of 000 followers. 
    no = len(names)
    thousands = pd.DataFrame(range(1,301))

    # gets the count of how many edges are in each thousand. 
    data = []
    for i in range(no):
        data.append(pd.DataFrame(eval(names[i] + '_edges.thousand.value_counts().sort_index()')))

    # plots the data. 
    fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
    axes.set_title(label)
    axes.set_xlabel('Last 300k followers (Oct 22)')
    axes.set_ylabel('Follows the opposition (per 000)')

    # creates a for loop to plot each party. 
    for i in range(no):
        axes.plot(thousands.iloc[:,0]*10**3, data[i], color = colors[i], label=names[i])
    
    axes.set_ylim(0, 500)
    plt.gca().invert_xaxis()
    plt.legend(loc="upper left")
    plt.show()

get_plot(names = ['dems', 'reps'], label = 'The US')
get_plot(names = ['tories', 'labour'], label = 'The UK')
```

```python
# creates variables for party no. and the no. of 000 followers. 
thousands = pd.DataFrame(range(1,301))

# gets the combined country count of how many edges in each thousand. 
us_combined = (dems_edges.thousand.value_counts().sort_index() + reps_edges.thousand.value_counts().sort_index()) / 2
uk_combined = (tories_edges.thousand.value_counts().sort_index() + labour_edges.thousand.value_counts().sort_index()) / 2

# plots the combined data. 
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
axes.set_title('Combined country averages')
axes.set_xlabel('Last 300k followers (Oct 22)')
axes.set_ylabel('Follows the opposition (per 000)')
axes.plot(thousands.iloc[:,0]*10**3, us_combined, label='The US')
axes.plot(thousands.iloc[:,0]*10**3, uk_combined, label='The UK')
plt.gca().invert_xaxis()
plt.legend(loc="upper left")
plt.show()
```

```python
# imports the math module. 
import math

# smoothens the data by taking an anverage of the observations closest to i. 
def smoothen_data(data, avg = 3):
    
    # gets the number of observations before and after i to average. 
    change = math.floor(avg/2)
    
    # averages the middle data. 
    for i in range(1, len(data)+1):
        try:
            if data[i-change] >= 0 and data[i+change] >= 0:
                try:
                    data[i] = data[i-change:(i+change)+1].sum() / avg
                except:
                    pass 
        except:
            
            # averages the start data.
            try:
                if data[i+change] >= 0:
                    backwards = []
                    for b in range(change):
                        try:
                            if data[i-b] >= 0:
                                backwards.append(b)
                        except:
                            pass   
                    b = max(backwards)
                    data[i] = data[i-b:i+change+(change-b)].sum() / avg 
                else:
                    pass
            except:
                pass
            
            # averages the end data.
            try:
                if data[i-change] >= 0:
                    forwards = []
                    for f in range(change):
                        try:
                            if data[i+f] >= 0:
                                forwards.append(f)
                        except:
                            pass 
                    f = max(forwards)
                    data[i] = data[(i-change-(change-f)):i+f].sum() / avg
                else:
                    pass
            except:
                pass
    return data

us_smooth = smoothen_data(us_combined, 120)
uk_smooth = smoothen_data(uk_combined, 120)

# plots the smoothened combined data. 
fig, axes = plt.subplots(nrows=1, ncols=1, figsize=(12,6))
axes.set_title('Smoothed averages')
axes.set_xlabel('Last 300k followers (Oct 22)')
axes.set_ylabel('Follows the opposition (per 000)')
axes.plot(thousands.iloc[:,0]*10**3, us_smooth, label='The US')
axes.plot(thousands.iloc[:,0]*10**3, uk_smooth, label='The UK')
plt.gca().invert_xaxis()
plt.legend(loc="upper left")
plt.show()
```

```python
# imports the mpatches module. 
import matplotlib.patches as mpatches

# visualizes a network.
def visualize_network(nodes, network_size, country, colors = ['red', 'blue']):
    
    options = {'node_color': 'black',
               'node_size': 0.1,
               'with_labels': False,
               'arrows': True,
               'alpha': 0.5,
               'linewidths': 0.0005,
               'width': 0.1,
               'font_size': 4}
    
    node_list = []
    for node in range(len(nodes)):
        node_list.append(eval(country + '.loc[' + country + '.party==\'' + nodes[node] + '\'].iloc[0:' + str(network_size) + ']'))

    g = nx.Graph()
    color_map = []
    for node in range(len(nodes)):
        for i in range(len(node_list[node])):
            g.add_edge(node_list[node].party.iloc[i], node_list[node].user_id.iloc[i])
            color_map.append(colors[node])
            
    plt.figure(1,figsize=(10,6))    
    nx.draw_networkx(g, edge_color=color_map, **options)
    
    node_1 = mpatches.Patch(color='blue', label=nodes[0])
    node_2 = mpatches.Patch(color='red', label=nodes[1])
    plt.legend(handles=[node_1, node_2], loc="upper left")
    
    plt.title('The ' + country.upper())
    plt.show()

visualize_network(nodes = ['dems', 'reps'], country = 'us', network_size = 2000)
visualize_network(nodes = ['tories', 'labour'], country = 'uk', network_size = 2000)
```
